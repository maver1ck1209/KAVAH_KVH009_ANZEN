```python
import cv2
import torch
from ultralytics import YOLO
import supervision as sv
```

```python
#live webcam
cap = cv2.VideoCapture(0)
model = YOLO("best.pt")
box_annotator = sv.BoxAnnotator(thickness=2,text_thickness=2,text_scale=1)
while True:
    ret, frame = cap.read()
    result = model(frame)[0]
    detections = sv.Detections.from_yolov8(result)
    labels = [f"{model.model.names[class_id]} {confidence:0.1f}" for _, confidence, class_id, _ in detections]
    frame = box_annotator.annotate(scene=frame, detections=detections, labels=labels)
    cv2.imshow("yolov8", frame)
    if (cv2.waitKey(1) == 27):
        cv2.destroyAllWindows()
        break
```

```python
import torch
import cv2
import numpy as np
import pyautogui as pg
from PIL import ImageGrab
from win32 import win32gui
from ultralytics import YOLO
import supervision as sv
```

```python
hwnd = win32gui.FindWindow(None, r'guns video from cctv - Google Search - Opera')
dimensions = win32gui.GetWindowRect(hwnd)
model = YOLO("best.pt")
box_annotator = sv.BoxAnnotator(thickness=2,text_thickness=2,text_scale=1)
while True:
    img = np.array(ImageGrab.grab(dimensions))
    img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    result = model(img)[0]
    detections = sv.Detections.from_yolov8(result)
    labels = [f"{model.model.names[class_id]} {confidence:0.1f}" for _, confidence, class_id, _ in detections]
    img = box_annotator.annotate(scene=img, detections=detections, labels=labels)
    cv2.imshow("yolov8", img)
    if (cv2.waitKey(1) == 27):
        cv2.destroyAllWindows()
        break
```
